```python
import whitebox
import rasterio
import matplotlib.pyplot as plt
```

```python
# Initialize the whitebox tools
wbt = whitebox.WhiteboxTools()

# Define input and output DEM file paths
input_dem = "/Users/monmita/Desktop/MonmitaMukherjee/Rajasthan_PSP_150125/Rajasthan_DEM.tif"  
output_fill_dem = "/Users/monmita/Desktop/MonmitaMukherjee/Rajasthan_PSP_150125/Rajasthan_Fill_DEM.tif" M

# Perform Fill DEM
try:
    print("Filling DEM to remove depressions...")
    wbt.fill_depressions(dem=input_dem, output=output_fill_dem)
    print("Fill DEM generated successfully!")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
# Visualize the input and output DEM
def plot_dem(file_path, title):
    with rasterio.open(file_path) as src:
        dem_data = src.read(1)
        plt.figure(figsize=(8, 6))
        plt.imshow(dem_data, cmap="terrain", extent=(src.bounds.left, src.bounds.right, src.bounds.bottom, src.bounds.top))
        plt.colorbar(label="Elevation (m)")
        plt.title(title)
        plt.xlabel("Longitude")
        plt.ylabel("Latitude")
        plt.show()
```

```python
# Plot input DEM
plot_dem(input_dem, "Input DEM")

# Plot output Fill DEM
plot_dem(output_fill_dem, "Filled DEM")
```
